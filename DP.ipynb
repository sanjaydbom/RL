Iterative policy evaluation for policy that randomly decided direction. (0,0) and (3,3) are termnal points. Each step has a reward of -1

```python
def show_board(array):
    for row in array:
        for val in row:
            precision = 5
            if(val <= -10):
                print(f"{val:.2f} ", end="")
            elif val < 0 or val >= 10:
                print(f"{val:.3f} ", end="")
            else:
                print(f"{val:.4f} ", end="")
        print()

v = [[0 for i in range(4)] for j in range(4)]
v_pred = [[0 for i in range(4)] for j in range(4)]

theta = 0.0001
gamma = 1
delta = 0.02
while delta > theta:
    delta = 0
    for i in range(4):
        for j in range(4):
            if (i == 0 and j == 0) or (i == 3 and j == 3):
                pass
            else:
                temp = 0
                temp += 0.25 * (-1 + gamma * v[i-1][j]) if i > 0 else 0.25 * (-1 + gamma * v[i][j])
                temp += 0.25 * (-1 + gamma * v[i+1][j]) if i < 3 else 0.25 * (-1 + gamma * v[i][j])
                temp += 0.25 * (-1 + gamma * v[i][j-1]) if j > 0 else 0.25 * (-1 + gamma * v[i][j])
                temp += 0.25 * (-1 + gamma * v[i][j+1]) if j < 3 else 0.25 * (-1 + gamma * v[i][j])
                if abs(v[i][j] - temp) > delta:
                    delta = abs(v[i][j]-temp)
                v_pred[i][j] = temp
    for i in range(4):
        for j in range(4):
            v[i][j] = v_pred[i][j]
            v_pred[i][j] = 0
show_board(v)
print()

delta = 0.02
while delta > theta:
    delta = 0
    for i in range(4):
        for j in range(4):
            if (i == 0 and j == 0) or (i == 3 and j == 3):
                pass
            else:
                temp = -1000
                if i > 0 and -1 + gamma * v[i-1][j] > temp:
                    temp = -1 + gamma * v[i-1][j]
                if i < 3 and -1 + gamma * v[i+1][j] > temp:
                    temp = -1 + gamma * v[i+1][j]
                if j > 0 and -1 + gamma * v[i][j-1] > temp:
                    temp = -1 + gamma * v[i][j-1]
                if j < 3 and -1 + gamma * v[i][j+1] > temp:
                    temp = -1 + gamma * v[i][j+1]
                if abs(v[i][j] - temp) > delta:
                    delta = abs(v[i][j]-temp)
                v_pred[i][j] = temp
    for i in range(4):
        for j in range(4):
            v[i][j] = v_pred[i][j]
            v_pred[i][j] = 0
show_board(v)
print()
```

Value Iteration. Greedy policy always moves to square with highest value. The output shows the evolution of the best policy. The final policy is optimal

```python
def reward(i,j):
    if i == 0 and j == 0:
        return 0
    elif i == 0 or (i == 2 and j != 3):
        return -10
    else:
        return -1

v = [[0 for i in range(4)] for j in range(4)]
v_pred = [[0 for i in range(4)] for j in range(4)]

theta = 0.01
gamma = 1
delta = 0.2
while delta > theta:
    delta = 0
    for i in range(4):
        for j in range(4):
            if (i == 0 and j == 0):
                pass
            else:
                temp = reward(i,j) * 0.25
                temp += 0.25 * gamma * v[i-1][j] if i > 0 else 0.25 * gamma * v[i][j]
                temp += 0.25 * gamma * v[i+1][j] if i < 3 else 0.25 * gamma * v[i][j]
                temp += 0.25 * gamma * v[i][j-1] if j > 0 else 0.25 * gamma * v[i][j]
                temp += 0.25 * gamma * v[i][j+1] if j < 3 else 0.25 * gamma * v[i][j]
                if abs(v[i][j] - temp) > delta:
                    delta = abs(v[i][j]-temp)
                v_pred[i][j] = temp
    for i in range(4):
        for j in range(4):
            v[i][j] = v_pred[i][j]
            v_pred[i][j] = 0
show_board(v)
print()
delta = 0.02
while delta > theta:
    delta = 0
    for i in range(4):
        for j in range(4):
            if i == 0 and j == 0:
                pass
            else:
                temp = -1000
                if i > 0 and reward(i,j) + gamma * v[i-1][j] > temp:
                    temp = reward(i,j) + gamma * v[i-1][j]
                if i < 3 and reward(i,j) + gamma * v[i+1][j] > temp:
                    temp = reward(i,j) + gamma * v[i+1][j]
                if j > 0 and reward(i,j) + gamma * v[i][j-1] > temp:
                    temp = reward(i,j) + gamma * v[i][j-1]
                if j < 3 and reward(i,j) + gamma * v[i][j+1] > temp:
                    temp = reward(i,j) + gamma * v[i][j+1]
                if abs(v[i][j] - temp) > delta:
                    delta = abs(v[i][j]-temp)
                v_pred[i][j] = temp
    for i in range(4):
        for j in range(4):
            v[i][j] = v_pred[i][j]
            v_pred[i][j] = 0
    show_board(v)
    print()
```
