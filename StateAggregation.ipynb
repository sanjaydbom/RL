```python
import random
import matplotlib.pyplot as plt
```

```python
weights = 0
alpha = 2e-5
for _ in range(10000):
    position = 500
    position_history = []
    while position > 0 and position < 1001:
        dir = random.choice([-1,1])
        steps = random.randint(1,100)
        position += dir * steps
        position_history.append(position)
    if position <= 0:
        reward = -1
    else:
        reward = 1

    position_history = position_history[:-1]
    for pos in position_history:
        weights += alpha * (reward - weights * (pos//100-5)) * (pos // 100 - 5 )
weights
```

```python
value_estimates = [weights * i for i in range(-5,6)]
x = [100 * i for i in range(0,11)]
plt.step(x, value_estimates, where='mid')
```

```python
true_ve = [weights * i for i in range(-500,501)]
true_x = [i for i in range(-500,501)]
plt.step(true_x, true_ve)
plt.show()
```
